```python
import os
from smosaic import mosaic, open_geojson
```

```python
path_dir = os.path.abspath("")
```

```python
geometry = open_geojson("luis-eduardo-magalhaes.json")
```

```python
stac_url = "https://data.inpe.br/bdc/stac/v1"
```

```python
result = mosaic(
    name="luis-eduardo-magalhaes",
    data_dir=os.path.join(path_dir),
    stac_url=stac_url,
    collection="S2_L2A-1", 
    geom=geometry,
    output_dir=os.path.join(path_dir, "output"),   
    mosaic_method="lcf", 
    start_year=2025,
    start_month=6,
    start_day=1,
    duration_months=3, 
    bands=["B02","B03","B04"]
)
```
